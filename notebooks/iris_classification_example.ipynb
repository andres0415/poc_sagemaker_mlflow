# Iris Classification - Basic ML Example

This notebook demonstrates a simple machine learning workflow using the Iris dataset.
Perfect for use with AWS SageMaker Studio and MLflow experiments.

## 1. Import Libraries

```python
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from data_loader import load_iris
from preprocessing import preprocess_data
from models import compare_models
from evaluation import evaluate_model, plot_confusion_matrix, compare_model_accuracies

print("Libraries imported successfully!")
```

## 2. Load Data

```python
# Load the Iris dataset
X, y = load_iris()

print(f"Dataset shape: {X.shape}")
print(f"Features: {list(X.columns)}")
print(f"Classes: {y.unique()}")
print(f"Class distribution:\n{y.value_counts()}")
```

## 3. Explore Data

```python
# Display basic statistics
print("Dataset Info:")
print(X.describe())

# Visualize data
plt.figure(figsize=(12, 8))
sns.pairplot(pd.concat([X, y], axis=1), hue='species')
plt.suptitle('Iris Dataset - Pairwise Feature Relationships', y=1.02)
plt.show()
```

## 4. Preprocess Data

```python
# Split and scale the data
X_train, X_test, y_train, y_test = preprocess_data(X, y, test_size=0.3, scale=True)

print(f"Training set: {X_train.shape}")
print(f"Test set: {X_test.shape}")
```

## 5. Train and Compare Models

```python
# Train and compare three basic models
results = compare_models(X_train, y_train, X_test, y_test)

print(f"\nBest performing model: {results['best_model']}")
```

## 6. Visualize Results

```python
# Plot model comparison
compare_model_accuracies(results)
```

```python
# Plot confusion matrix for the best model
best_model_name = results['best_model']
best_model = results[best_model_name]['model']
y_pred = results[best_model_name]['predictions']

plot_confusion_matrix(y_test, y_pred, best_model_name)
```

## 7. Model Evaluation Summary

```python
# Detailed evaluation of the best model
evaluate_model(best_model, X_test, y_test, best_model_name)
```
